# Cleanup all the data so it can be used in regression before analysis


```python
train['loan_status'] = train['loan_status'].map({'Charged Off': 1, 'Fully Paid': 0})
train['term'] = train['term'].map({' 36 months': 36, ' 60 months': 60, '36 months': 36, '60 months': 60})
train['emp_length'] = train['emp_length'].map({'10+ years': 10,'1 year': 1,'3 years': 3,'8 years': 8,'9 years': 9,'5 years': 5,'< 1 year': 0,'4 years': 4,'6 years': 6,'2 years': 2,'7 years': 7,'n/a': np.NaN})
train['int_rate'] = train.int_rate.str.replace('%','')
train['sub_grade'] = train['sub_grade'].map({'A1': 1,'A2': 2,'A3': 3,'A4': 4,'A5': 5,'B1': 6,'B2': 7,'B3': 8,'B4': 9,'B5': 10,'C1': 11,'C2': 12,'C3': 13,'C4': 14,'C5': 15,'D1': 16,'D2': 17,'D3': 18,'D4': 19,'D5': 20,'E1': 21,'E2': 22,'E3': 23,'E4': 24,'E5': 25,'F1': 26,'F2': 27,'F3': 28,'F4': 29,'F5': 30,'G1': 31,'G2': 32,'G3': 33,'G4': 34,'G5': 35})
train['revol_util'] = train.revol_util.str.replace('%','')
rvalue = r"[A-Z][a-z][a-z][-]"
train['earliest_cr_line'] = train.earliest_cr_line.str.replace(rvalue,'')
```

```python
train = train.rename(index=str, columns={"loan_status": "target"})
```

```python
train
```

```python
data = []
for f in train.columns:
    #print(f, train[f].dtype)
    # Defining the role
    if f == 'target':
        role = 'target'
    elif f == 'id':
        role = 'id'
    else:
        role = 'input'
         
    # Defining the level
    if 'bin' in f or f == 'target':
        level = 'binary'
    elif 'cat' in f or f == 'id':
        level = 'nominal'
    elif train[f].dtype == 'int64':
        level = 'ordinal'    
    elif train[f].dtype == 'float64':
        level = 'interval'

        
    # Initialize keep to True for all variables except for id
    keep = True
    if f == 'id':
        keep = False
    
    # Defining the data type 
    dtype = train[f].dtype
    
    # Creating a Dict that contains all the metadata for the variable
    f_dict = {
        'varname': f,
        'role': role,
        'level': level,
        'keep': keep,
        'dtype': dtype
    }
    data.append(f_dict)
    
meta = pd.DataFrame(data, columns=['varname', 'role', 'level', 'keep', 'dtype'])
meta.set_index('varname', inplace=True)
```

```python
#Risk Modeling
RiskTolerance = .7
RiskStep = .01
AmountAvailable = 468.07
InvestmentNumber = int(AmountAvailable/50)
AnalysisRiskReturn = pd.DataFrame(columns=['RiskTolerance', 'RiskScore', 'ExpReturn', 'ExpValue', 'StdPortfolio', 'Meanint_rate', 'MeanChargeoff', 'SumChargeoff'])

while RiskTolerance < 1.0:
    Output = Invest.join(pd.DataFrame(logistic.predict_proba(InvestAnalysis)))
    Output['IntAdjRiskReturn'] = Output[0]* Output['int_rate']
    Output = Output[Output[0] > RiskTolerance]
    Output = Output.sort_values(by=['IntAdjRiskReturn'], ascending=False)
    Output = Output.head(InvestmentNumber)
    #print('Risk Tolerance Level :', RiskTolerance, '| Risk Score: ', round(Output[1].sum(),4), '| Expected Return: ', round((Output['IntAdjRiskReturn'].mean()) ,4), '| Expected Value: $', round((Output['IntAdjRiskReturn'].mean()/100 * AmountAvailable),4))
    if Output.shape[0] == InvestmentNumber:
        AnalysisRiskReturn = AnalysisRiskReturn.append({'RiskTolerance': RiskTolerance,'RiskScore': round(Output[1].sum(),4), 'ExpReturn': round((Output['IntAdjRiskReturn'].mean()) ,4), 'ExpValue': round((Output['IntAdjRiskReturn'].mean()/100 * (Output.shape[0]*50)),4), 'StdPortfolio': Output['IntAdjRiskReturn'].std(), 'Meanint_rate': Output['int_rate'].mean(),'MeanChargeoff': Output[1].mean(),'SumChargeoff': Output[1].sum()}, ignore_index=True)
    RiskTolerance += RiskStep
    
    if Output['IntAdjRiskReturn'].mean() < .99:
        break



    


```

```python
AnalysisRiskReturn = AnalysisRiskReturn.sort_values(by=['ExpReturn'], ascending=False)
# importing the required module 
import matplotlib.pyplot as plt 
  
# plotting the points  
plt.plot(AnalysisRiskReturn['ExpReturn'], AnalysisRiskReturn['SumChargeoff']) 
  
# naming the x axis 
plt.xlabel('ExpReturn') 
# naming the y axis 
plt.ylabel('SumChargeoff') 
  
# giving a title to my graph 
plt.title('Frontier Analysis') 
  
# function to show the plot 
plt.show() 

AnalysisRiskReturn = AnalysisRiskReturn.sort_values(by=['ExpReturn'], ascending=False)

# plotting the points  
plt.plot(AnalysisRiskReturn['ExpReturn'], AnalysisRiskReturn['RiskScore']) 

# naming the x axis 
plt.xlabel('ExpReturn') 
# naming the y axis 
plt.ylabel('RiskScore') 
  
# giving a title to my graph 
plt.title('Frontier Analysis') 
  
# function to show the plot 
plt.show() 


AnalysisRiskReturn = AnalysisRiskReturn.sort_values(by=['ExpReturn'], ascending=True)
# plotting the points  
plt.plot(AnalysisRiskReturn['ExpReturn'], AnalysisRiskReturn['RiskTolerance']) 

# naming the x axis 
plt.xlabel('ExpReturn') 
# naming the y axis 
plt.ylabel('RiskTolerance') 
  
# giving a title to my graph 
plt.title('Frontier Analysis') 
  
# function to show the plot 
plt.show() 

```
